```python
conn = sqlite3.connect(':memory:')

df_l2a.to_sql('l2a', conn, index=False)
df.to_sql('fire', conn, index=False)
```

```python
qry = '''
SELECT * FROM l2a
JOIN fire ON
SENSING_TIME BETWEEN STARTED AND ENDED
WHERE l2a.MGRS_TILE = fire.MGRS
'''
```

```python
df_merge = pd.read_sql_query(qry, conn)
```

```python
# the sql join appeard to remove the datatime conversion
df_merge['SENSING_TIME'] = pd.to_datetime(df_merge['SENSING_TIME'])
```

```python
df_merge.info()
```

```python
#add date column
df_merge['DATE'] = df_merge['SENSING_TIME'].dt.date 
```

```python
'''
the below code eliminates duplicate .safe folders by saving the .safe folder that is the largest size.
throughout this research I found that there were often duplicates of one image scene and the quality varried, 
generally the larger size .safe folder had thhe best quality scene
https://stackoverflow.com/questions/12497402/python-pandas-remove-duplicates-by-columns-a-keeping-the-row-with-the-highest
'''
df_merge = df_merge.sort_values('TOTAL_SIZE', ascending=False).drop_duplicates(['MGRS_TILE', 'DATE']).sort_index()
```

```python
df_merge.info()
```

```python
print(len(df_merge['GRANULE_ID'].unique()))
print(len(df_merge['PRODUCT_ID'].unique()))
```

```python
'''
This finds the next granule id. Basically we want to feed sequntial images to our neural network and we do
this by tracking the granule id and the next tile granule id.
'''

# gets the next tile and time of the next tile grouped by MGRS
df_merge['NEXT_TILE'] = df_merge.sort_values('SENSING_TIME', ascending=False).groupby('MGRS')['GRANULE_ID'].shift()
df_merge['NEXT_TILE_TIME'] = df_merge.sort_values('SENSING_TIME', ascending=False).groupby('MGRS')['SENSING_TIME'].shift()
```

```python
# days betweeen first tile and next tile
# https://stackoverflow.com/questions/37840812/pandas-subtracting-two-date-columns-and-the-result-being-an-integer/46966942
df_merge['DAYS_BETWEEN'] = (df_merge['NEXT_TILE_TIME'] - df_merge['SENSING_TIME']).dt.days
```

```python
#remove tiles more than 10 days apart
df_merge[df_merge['DAYS_BETWEEN'] <= 10].info()
```

```python
df_merge =  df_merge[df_merge['DAYS_BETWEEN'] <= 10].sort_values('MGRS')
```

```python
# remove .safe files that are smaller than 300 mb
df_merge = df_merge[df_merge['TOTAL_SIZE'] > 300000000]
```

```python
#save progress
df_merge.to_csv('in_progess_2016.csv', index=False)
```
